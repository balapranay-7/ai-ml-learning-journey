# **List**

A list is an ordered,mutable (changeable) collection of itemsn>List can hold elements of data types, but usually contain similar items

# **Key features**


*   Ordered(items have a fixed order)
* Mutable (can change after creation )
*  Allows duplicates



```python
# Creating a list
fruits = ['apple','banana','cherry']
print(fruits)
#Indexing
print(fruits[0]) # first item
# slicing
print(fruits[0:2]) # first two items
# Common methods
fruits.append('orange')
print(fruits)
fruits.remove('banana')
print(fruits)
fruits.sort()
print(fruits)
```

#**Sets**
A **set** is an unordered collection of unique items .It does not allow duplicates
**Key Features**
* Unordered(no indexing)
* No duplicates
* Mutable (can be added or removed elements )


```python
# Creating a  set
unique_numbers = {1,2,3,4,5}
print(unique_numbers)
# Adding elements
unique_numbers.add(5)
print(unique_numbers)
# Set operations
a = {1,2,3}
b = {3,4,5}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))

```

```python
india_1 = {"hampi","taj mahal","mysore palace","bangalore fort"}
india_2 = {"mysore palace" , "bangalore fort","japur","nagaland","sevenssisters"}
print(india_1.union(india_2))
print(india_1.intersection(india_2))
print(india_1.difference(india_2))

```

#**Dictionaries**
A dictionary is collection of key-value pairs.Each key must be unique ,and is used to access its associated value

**Key Features**
* Unordered
* Keys must be unique and immutable
* Values can be any type





```python
#Creating a dictionaary
student = {
    'name':'Pranay',
    'age':20,
    'score':80

}
print(student)
#Access values
print(student['name'])
#Adding a new key-value pair
student['grade'] = 'A'
print(student)
```

```python
student  ={
    'name':{'Alice ','sho'},
    'age':{20,22},
    'score':{80,24},
}
print(student)
```
