# Simple Calculator

```python
def calculator(a,b,op):
    if op == '+':
        return a + b
    elif op =='-':
        return a - b
    elif op =='*':
      return a * b
    elif op == a / b:
      return a / b
    else:
       return "Invalid operator"
print(calculator(10,5,'/'))
print(calculator(10,5,'+'))
```
