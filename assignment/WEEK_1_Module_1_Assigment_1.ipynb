2. Convert the string "12345" into: • an integer • a float Show the result of adding 5 to both.

```python
num_str = "12345"

num_int = int(num_str)
num_float = float(num_str)

print("Integer:", num_int)
print("Integer + 5 =", num_int + 5)

print("Float:", num_float)
print("Float + 5 =", num_float + 5)
```

3. Write a program that takes two numbers as input and prints: • Their sum • Their difference • Their product • Their division

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)

if num2 != 0:
    print("Division =", num1 / num2)
else:
    print("Division by zero is not allowed.")
```

4. Write a program to check whether a given number is positive, negative, or zero.

```python
number = float(input("Enter a number: "))

if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```
